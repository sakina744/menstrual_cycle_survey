```python
#Data Preprocessing
#Rescaling
import pandas as pd
import numpy as np
from sklearn import preprocessing
import scipy.stats as s
print("Data Scaled between 0 t0 1")
df_age=pd.DataFrame(df['Age:'])
rescaling= preprocessing.MinMaxScaler(feature_range=(0,5))
rescaled=rescaling.fit_transform(df_age)
print("Min Max Scaled Data")
print("----------------------------")
print(rescaled.round(2))
```

```python
#normalizing
dn=preprocessing.normalize(df_age, norm='l1')
print("L1 Normalize Data ")
print("----------------------------------------")
print(dn.round(2))
```

```python
#Binarizing
data_binarized=preprocessing.Binarizer(threshold=19).transform(df_age)
print("Binarized data")
print("-----------------------------------")
print(data_binarized)
```

```python
#standarizing Data
print("Standardizing Data")
print("-------------------------------------")
print("Original Data\n",df_age)
print("Initial Mean:",s.tmean(df_age).round(2))
print("Initial Standard Deviation:",round(df_age.std(),2))
df_age_scaled=preprocessing.scale(df_age)
df_age_scaled.mean(axis=0)
df_age_scaled.std(axis=0)
print("Standardized Data\n",df_age_scaled.round(2))
print("Scaled Mean:",s.tmean(df_age_scaled).round(2))
print("Scaled Standard Deviation:",round(df_age_scaled.std(),2))
```

```python
#normal distribution
import numpy as np
import matplotlib.pyplot as plt
x=list(df["Hygiene and health"])
def normal_dist(x, mean, sd):
    pro=(np.pi*sd)*np.exp(-0.5*((x-mean)/sd)**2)
    return pro
mean= np.mean(x)
sd=np.std(x)
pdf=normal_dist(x,mean,sd)

plt.plot(x,pdf, color='red')
plt.title("Hygiene and health")
plt.show()
```
